```python
import pandas as pd
import numpy as np

data = {
    "Name": ["Pavan", "Rahul", "Kiran", "Sai"],
    "Marks": [95, np.nan, 91, 88],
    "Age": [18, 19, np.nan, 20]
}

df = pd.DataFrame(data)

print(df)
```

```python
print(df.isnull())
print(df.isnull().sum())
print(df.dropna())
print(df.fillna(0))
```

```python
df["Marks"]=df["Marks"].fillna(df["Marks"].mean())
print(df)
df["Age"]=df["Age"].fillna(df["Age"].mode())
print(df)
```

```python
import pandas as pd

data = {
    "Name":["Pavan","Rahul","Kiran","Sai","Ramesh"],
    "Branch":["AIML","CSE","AIML","CSE","AIML"],
    "Marks":[95,88,91,84,97]
}

df = pd.DataFrame(data)

print(df)
```

```python
print(df.groupby("Branch"))
print(df.groupby("Branch")["Marks"].mean())
print(df.groupby("Branch")["Marks"].max())
print(df.groupby("Branch")["Marks"].min())
```

```python
print(df.groupby("Branch")["Marks"].agg(["mean","max","min","count","median"]))
```

```python
import pandas as pd

students = pd.DataFrame({
    "Roll": [101,102,103,104],
    "Name": ["Pavan","Rahul","Kiran","Rohan"]
})

marks = pd.DataFrame({
    "Roll": [101,102,103],
    "Marks": [95,88,91]
})
```

```python
result=pd.merge(students,marks,on="Roll")
print(result)
left=pd.merge(students,marks,on="Roll",how="left")
print(left)
outer=left=pd.merge(students,marks,on="Roll",how="outer")
print(outer)
```

```python
import pandas as pd

data = {
    "Name": ["Pavan", "Rahul", "Pavan"],
    "Marks": [95, 88, 95]
}

df = pd.DataFrame(data)

print(df)
```

```python
print(df.drop_duplicates())
df.rename(columns={"Marks":"Score"},inplace=True)
print(df)
df["Score"]=df["Score"].astype(float)
print(df.dtypes)
df["Name"]=df["Name"].replace("Rahul","Ravi")
print(df)
```
