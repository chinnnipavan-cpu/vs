```python
class Account:
    def __init__(self,bal,acc):
        self.bal=bal
        self.acc=acc
        #debit method
    def debit(self,amount):
        self.bal-=amount
        print(f"Rs {amount} was debited ={self.get_bal()}")
    def credit(self,am):
        self.bal+=am
        print(f"The amount was credited is {am} remaining {self.get_bal()}")
    def get_bal(self):
        return self.bal


    

acc1=Account(100000,1253248)
acc1.credit(1987)
acc1.debit(1987)
acc1.get_bal()
```

```python
class Car:
    def __init__(self,type):
        self.type=type
    @staticmethod
    def start():
        print("car has started")
    @staticmethod
    def stop():
        print("car has stopped")
class BMW(Car):
    def __init__(self,brand,type):
        self.brand=brand
        super().__init__(type)
        super().start()
car1=BMW("M4","rich")
print(car1.type)
```

```python
class Complex:
    def __init__(self,real,img):
        self.real=real
        self.img=img
    def shownum(self):
        print(self.real,"+",self.img,"j")
        #dunder function
    def __add__(self,n2):
        newreal=self.real+n2.real
        imgreal=self.img+n2.img
        return Complex(newreal,imgreal)
    def __sub__(self,n2):
        newreal=self.real-n2.real
        imgreal=self.img-n2.img
        return Complex(newreal,imgreal)
n1=Complex(1,3)
n2=Complex(3,5)
n1.shownum()
n2.shownum()
n3=n1+n2
n4=n1-n2
n3.shownum()
n4.shownum()
```

```python
#proj1
class Circle:
    def __init__(self,radius):
        self.radius=radius
    def Area(self):
        return 3.14*self.radius*self.radius
    def Perim(self):
        return 2*3.14*self.radius
c1=Circle(25)
print(c1.Area())
print(c1.Perim())
```

```python
#proj2
class Employee:
    def __init__(self,role,dep,sal):
        self.role=role
        self.dep=dep
        self.sal=sal
    def show(self):
        print(f"The emplyee role is{self.role} and dep is {self.dep} and sal is {self.sal}")
class Engineer(Employee):
    def __init__(self,name,age):
        self.name=name
        self.age=age
        super().__init__("Engineer","ai",10000000000000)
    
eng1=Engineer("pavan",20)
eng1.show()
```
