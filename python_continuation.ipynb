```python
def main():
    num=get_num()
    meow(num)
def get_num():
    while True:
        n=int(input("Enter a number: "))
        if n>0:
            return n
def meow(n):
    for _ in range(n):
        print("meow")
main()
```

```python
def main():
    print_row(4)
def print_row(n):
    for _ in range(n):
        for _ in range(n):
            print("*",end="")
        print()
main()
```
