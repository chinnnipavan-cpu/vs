## 1. Create Arrays from Python Lists

NumPy arrays can be created from Python lists using `np.array()`.

```python
import numpy as np

arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([[1, 2, 3], [4, 5, 6]])
arr3 = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])

print(arr1)
print(arr2)
print(arr3)
```


````python
# Create 1D, 2D, and 3D arrays
arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([[1, 2, 3], [4, 5, 6]])
arr3 = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])

print(arr1)
print(arr2)
print(arr3)
````

## 2. Array Attributes and Data Types

You can inspect an array using attributes such as shape, size, ndim, dtype, and itemsize.

```python
arr = np.array([[1, 2, 3], [4, 5, 6]], dtype=float)
print(arr.shape)
print(arr.size)
print(arr.ndim)
print(arr.dtype)
print(arr.itemsize)
```


````python
arr = np.array([[1, 2, 3], [4, 5, 6]], dtype=float)
print("shape:", arr.shape)
print("size:", arr.size)
print("ndim:", arr.ndim)
print("dtype:", arr.dtype)
print("itemsize:", arr.itemsize)
````

## 3. Indexing, Slicing, and Fancy Indexing

You can access elements using simple indexes, slices, boolean masks, and integer arrays.

```python
arr = np.array([10, 20, 30, 40, 50])
print(arr[0])
print(arr[1:4])
print(arr[arr > 30])
```


````python
arr = np.array([10, 20, 30, 40, 50])
print(arr[0])
print(arr[1:4])
print(arr[arr > 30])

mat = np.array([[1, 2, 3], [4, 5, 6]])
print(mat[1, 2])
print(mat[:, 1])
````

## 4. Reshape, Resize, and Flatten Arrays

Use reshape(), ravel(), flatten(), transpose(), and expand_dims() to change array structure.

```python
arr = np.arange(1, 10)
print(arr.reshape(3, 3))
print(arr.ravel())
print(arr.reshape(1, -1).shape)
```


````python
arr = np.arange(1, 10)
print(arr.reshape(3, 3))
print(arr.ravel())
print(arr.reshape(1, -1).shape)
print(arr.T if arr.ndim > 1 else arr)
````

## 5. Vectorized Operations and Broadcasting

NumPy performs elementwise operations efficiently and supports broadcasting across shapes.

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print(a + b)
print(a * 2)
```


````python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print(a + b)
print(a * 2)

x = np.array([[1, 2, 3], [4, 5, 6]])
y = np.array([1, 0, 1])
print(x + y)
````

## 6. Aggregation and Statistical Computations

NumPy provides fast functions for computing statistics over arrays.

```python
arr = np.array([1, 2, 3, 4, 5])
print(np.sum(arr))
print(np.mean(arr))
print(np.std(arr))
```


````python
arr = np.array([1, 2, 3, 4, 5])
print("sum:", np.sum(arr))
print("mean:", np.mean(arr))
print("std:", np.std(arr))
print("min:", np.min(arr))
print("max:", np.max(arr))
````

## 7. Boolean Masking and Conditional Selection

Boolean masks help filter arrays based on conditions.

```python
arr = np.array([1, 2, 3, 4, 5, 6])
mask = arr > 3
print(mask)
print(arr[mask])
```


````python
arr = np.array([1, 2, 3, 4, 5, 6])
mask = arr > 3
print(mask)
print(arr[mask])
print(arr[arr % 2 == 0])
````

## 8. Stacking, Splitting, and Concatenation

NumPy can combine arrays in several useful ways.

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print(np.concatenate([a, b]))
print(np.vstack([a, b]))
```


````python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print(np.concatenate([a, b]))
print(np.vstack([a, b]))
print(np.hstack([a, b]))
print(np.split(np.array([1, 2, 3, 4, 5, 6]), 3))
````

## 9. Random Numbers and Basic Distributions

NumPy can generate random values for simulations and sampling.

```python
rng = np.random.default_rng(42)
print(rng.integers(0, 10, size=5))
print(rng.random((2, 3)))
```


````python
rng = np.random.default_rng(42)
print(rng.integers(0, 10, size=5))
print(rng.random((2, 3)))
print(rng.normal(loc=0.0, scale=1.0, size=5))
````
